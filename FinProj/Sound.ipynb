```python
import matplotlib.pyplot as plt
import scipy as sci
import numpy as np

# 1. Define time parameters
sampling_rate = 44100  # Samples per second
duration = 10.0        # Total time in seconds
t = np.linspace(0, duration, int(sampling_rate * duration), endpoint=False)
t2 = np.linspace(0, duration, int(sampling_rate * duration), endpoint=False)

# 2. Define frequencies (in Hz)
f1 = 400   # 2 cycles per second
f2 = 402   # 5 cycles per second
f3 = 407  # 10 cycles per second
f4 = 408

# 3. Generate sine waves
wave_1 = np.sin(2 * np.pi * f1 * t2)
wave_2 = np.sin(2 * np.pi * f2 * t2)
wave_3 = np.sin(2 * np.pi * f3 * t2)
wave_4 = np.sin(2 * np.pi * f4 * t2)

beat12 = np.cos(2 * np.pi * np.abs(f1-f2) * t)
beat13 = np.cos(2 * np.pi * np.abs(f1-f3) * t)
beat23 = np.cos(2 * np.pi * np.abs(f2-f3) * t)
beat14 = np.cos(2 * np.pi * np.abs(f1-f4) * t)
beat24 = np.cos(2 * np.pi * np.abs(f2-f4) * t)
beat34 = np.cos(2 * np.pi * np.abs(f3-f4) * t)

beatsum = beat12 + beat13 + beat23 + beat14 + beat24 + beat34

# # 4. Plot the waves
plt.figure(figsize=(10, 6))
# plt.plot(t2, wave_1, label=f'{f1} Hz')
# plt.plot(t, wave_2, label=f'{f2} Hz')
# plt.plot(t, wave_3, label=f'{f3} Hz')

# plt.title('Sine Waves at Different Frequencies')
# plt.xlabel('Time (seconds)')
# plt.ylabel('Amplitude')
# plt.legend()
# plt.grid(True)
# plt.show()

# Summing the waves from the previous example
composite_wave = wave_1 + wave_2 + wave_3 + wave_4

# plt.figure(figsize=(10, 4))
plt.plot(t2, composite_wave, color='orange')
plt.plot(t, beatsum, color='black')
# plt.plot(t, beat12, color='green')
# plt.plot(t, beat13, color='blue')
# plt.plot(t, beat23, color='purple')
#plt.plot(t, np.abs(np.sin(2*np.pi*15*t)))
plt.title('Combined Sine Wave (400Hz + 420Hz + 440Hz)')
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.xlim(0,4)
plt.show()
```
