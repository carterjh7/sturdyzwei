```python
data["E_lab"] = data["E1"] + data["E2"]
data["px_lab"] = data["px1 "] + data["px2"]
data["py_lab"] = data["py1"] + data["py2"]
data["pz_lab"] = data["pz1"] + data["pz2"]
data["inMass"] = (data["E_lab"]**2 - (data["px_lab"]**2 + data["py_lab"]**2 + data["pz_lab"]**2))**0.5
```

```python
import matplotlib.pyplot as plt
import numpy as np

binning = np.linspace(2,4, 100)

filter_data = data[data["Q1"] != data["Q2"]]
filter_data_same = data[data["Q1"] == data["Q2"]]
plt.hist(filter_data["inMass"], bins = binning)
#plt.hist(filter_data_same["inMass"], bins = 1000)
plt.grid()
plt.xlim(2, 4)
plt.show()
```
