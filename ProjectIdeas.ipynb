In this project we decided to investigate how beat frequency is determined when 3 or more frequencies are present.

```python
import matplotlib.pyplot as plt
import numpy as np


t = np.arange(0, 12*np.pi, 0.1)

#Function is the superposition of 3 sine functions with different frequencies
func1 = np.sin(t) + np.sin(2*t) + np.sin(1.5*t)

plt.plot(t, func1)
#plt.axis((-1,7,-1,1))

#Beat period appears to be about 4 pi
```

```python
func2 = np.sin(t) + np.sin(2*t) + np.sin(3*t)

plt.plot(t, func2)
#plt.axis((-1,7,-1,1))

#Beat period appears to be about 2 pi
```

```python
#Func 3 is just function 2 but scaled so that each frequency is twice the frequency present in func2
func3 = np.sin(2*t) + np.sin(4*t) + np.sin(6*t)

plt.plot(t, func3)
```
