```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
fs = 44100          # Sampling rate (Hz)
duration = 1.0      # Seconds
freq = 440          # Frequency of the sine wave (Hz)

# Time array
t = np.linspace(0, duration, int(fs * duration), endpoint=False)

# Sine wave
signal = np.sin(2 * np.pi * freq * t)
sig2 = np.sin(2*np.pi*(freq + 20)*t)
sig3 = np.sin(2*np.pi*(freq - 20)*t)

# Plot
plt.plot(t[:1000], signal[:1000])  # Plot first 1000 samples for clarity
plt.plot(t[:1000], sig2[:1000])
plt.plot(t[:1000], sig3[:1000])
plt.title("440 Hz Sine Wave")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.show()

sup = signal + sig2 + sig3
```

```python
ad(sup, rate = 44100)
```

```python
plt.plot(t[:5000], sup[:5000]);
```

```python
import scipy
ft = scipy.fft.rfft(sup)
plt.xlim(300,500)
plt.plot(np.abs(ft));
```
